```python
from matplotlib import pyplot as plt#Интегрируем библиотеку
# Список векторов, определяющих динозавра
dino_vectors = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4),

                (-5, 3), (-5, 2), (-2, 2), (-5, 1), (-4, 0), (-2, 1),

                (-1, 0), (0, -3), (-1, -4), (1, -4), (2, -3), (1, -2),

                (3, -1), (5, 1)]
# Разделяем координаты x и y
x_coords, y_coords = zip(*dino_vectors)
# Построение графика
plt.figure(figsize=(8, 6))#Создаем новое окно для графика
plt.scatter(x_coords, y_coords, color="blue", label="Точки динозавра")#Строим рассеивание точек
plt.title("Точки контура динозавра")#Устанавливаем заголовок графика
plt.xlabel("X-координаты")#Подпись для оси Х
plt.ylabel("Y-координаты")#Подпись для оси У
plt.grid(True)#Включаем сетку для удобства восприятия
plt.legend()#Показываем легенду графика
plt.show()#Отображаем график
```

```python
# Добавляем соединяющие линии
plt.figure(figsize=(8, 6))#Создаем новое окно для графика
plt.plot(*zip(*dino_vectors), linestyle="-", color="green", label="Контур динозавра")#Строим линию,соединящую точки векторов
plt.scatter(x_coords, y_coords, color="red", label="Точки динозавра")#Добавляем точки динозавра на график
plt.title("Контур динозавра с точками")#Устанавливаем заголовок
plt.xlabel("X-координаты")#Подпись оси Х
plt.ylabel("Y-координаты")#Подпись для оси У
plt.grid(True)#Включаем сетку
plt.legend()#Отобржаем легенду
plt.show()#Показываем конечный график
```

```python
from math import radians, cos, sin#Импорт необходимой библиотеки
# Функция вращения
def rotate(vector, theta):
    angle = radians(theta)
    return (vector[0] * cos(angle) - vector[1] * sin(angle),
            vector[0] * sin(angle) + vector[1] * cos(angle))
# Вращение фигуры
theta = 45
rotated_dino = [rotate(v, theta) for v in dino_vectors]
# Визуализация
plt.figure(figsize=(8, 6))#Создаем новое окно для графика и задаем его размеры
plt.plot(*zip(*dino_vectors), linestyle="-", color="blue", label="Исходный динозавр")#Рисуем повернутого динозавра
plt.plot(*zip(*rotated_dino), linestyle="--", color="red", label=f"Динозавр, вращённый на {theta}°")#
plt.title("Вращение динозавра")#Устанавливаем заголовок
plt.legend()#Отображаем легенду графика
plt.grid(True)#Включаем отображение сетки на графике
plt.show()#Отображаем график
```

```python
import numpy as np#Импорт необходимой библиотеки
x = np.arange(-10, 11, 1)#Создаем массив,содержащий значения от -10 до 10 с шагом 1
y = x**2#Вычисляем значение у
plt.figure(figsize=(8, 6))#Создаем новое окно для графика
plt.plot(x, y, marker="o", color="purple", label="y = x^2")#
plt.title("График функции y = x^2")#Устанавливаем заголовок графика
plt.xlabel("X")#Устанавливаем подписи для оси х
plt.ylabel("Y")#Устанавливаем подписи для оси у
plt.axhline(0, color="black",linewidth=0.5)#Добавляем вертикальную линию
plt.axvline(0, color="black",linewidth=0.5)#
plt.grid(True)#Включаем отображение сетки на графике для улучшения чтения значений
plt.legend()#Отображаем легенду графика
plt.show()#Отображаем график
```

```python
def shift_vectors(vectors, dx, dy):#Сдвигает каждый вектор на заданные значения
    return [(x + dx, y + dy) for x, y in vectors]
shifted_dino = shift_vectors(dino_vectors, -2, -2)#Сдвигаем координаты динозавра на -2 по оси X и -2 по оси Y
# Отображение исходного и сдвинутого динозавров
plt.figure(figsize=(8, 6))#Создаем новое окно для графика с размером 8x6
plt.plot(*zip(*dino_vectors), linestyle="-", color="blue", label="Исходный динозавр")#Рисуем исходного динозавра
plt.plot(*zip(*shifted_dino), linestyle="--", color="orange", label="Сдвинутый динозавр")#Рисуем сдвинутого динозавра
plt.scatter(*zip(*dino_vectors), color="blue")#Отображаем исходного динозавра точками
plt.scatter(*zip(*shifted_dino), color="orange")#Отображаем сдвинутого динозавра точками
plt.title("Перемещение динозавра")#Устанавливаем заголовок графика
plt.xlabel("X")#Подписываем ось х
plt.ylabel("Y")#Подписываем ось у
plt.grid(True)#Включаем сетку на графике
plt.legend()#Отображаем легенду
plt.show()#Отображаем график
```

```python
# Функция для создания сетки из динозавров
def create_dino_grid(vectors, grid_size=5, spacing=12):
    # Принимает список векторов, представляющих контур динозавра (vectors),
    # размер сетки (grid_size, по умолчанию 5) и расстояние между динозаврами (spacing, по умолчанию 12).

    dinos = [] # Создаем пустой список для хранения всех динозавров в сетке.

    # Внешний цикл перебирает строки сетки.
    for i in range(-grid_size, grid_size):
        # Внутренний цикл перебирает столбцы сетки.
        for j in range(-grid_size, grid_size):
            # Для каждой позиции в сетке сдвигаем исходного динозавра на соответствующее расстояние по осям x и y,
            # используя функцию shift_vectors (которая предполагает, что она уже определена).
            # i * spacing - величина сдвига по оси x.
            # j * spacing - величина сдвига по оси y.
            dinos.append(shift_vectors(vectors, i * spacing, j * spacing))

    return dinos # Возвращаем список, содержащий все сдвинутые контуры динозавров.

# Создаем сетку из динозавров, используя функцию create_dino_grid.
grid_dinos = create_dino_grid(dino_vectors)

# Построение всех динозавров
# Создаем фигуру для графика.
plt.figure(figsize=(10, 10))

# Перебираем все контуры динозавров в сетке.
for dino in grid_dinos:
    # Отображаем каждый контур динозавра на графике.
    # *zip(*dino) разделяет координаты x и y из списка dino.
    # linestyle="-" задает стиль линии как сплошную.
    # color="#997a8d" задает цвет линии (серо-фиолетовый).
    plt.plot(*zip(*dino), linestyle="-", color="#997a8d")

# Добавляем заголовок к графику.
plt.title("Сетка из динозавров")

# Добавляем подписи к осям x и y.
plt.xlabel("X")
plt.ylabel("Y")

# Задаем цвет фона графика (светло-серый).
plt.gca().set_facecolor("#dad4d8")

# Отключаем отображение сетки на графике (чтобы не загромождать его).
plt.grid(False)

# Отображаем график.
plt.show()
```

```python
# Функция сложения векторов
# Определяет функцию add, которая принимает два вектора v1 и v2 в качестве аргументов.
def add(v1, v2):
    # Возвращает новый вектор, представляющий собой сумму векторов v1 и v2.
    # Сложение выполняется поэлементно: первая координата v1 складывается с первой координатой v2,
    # вторая координата v1 складывается со второй координатой v2.
    return (v1[0] + v2[0], v1[1] + v2[1])

# Исходные векторы
# Определяет два вектора v1 и v2.
v1 = (4, 3) # Вектор v1 имеет координаты (4, 3).
v2 = (-1, 1) # Вектор v2 имеет координаты (-1, 1).
v_sum = add(v1, v2) # Вызывает функцию add для сложения векторов v1 и v2 и сохраняет результат в переменной v_sum.

# Построение векторов
# Создает новую фигуру для графика.
plt.figure(figsize=(6, 6)) # Задает размер фигуры (ширина 6 дюймов, высота 6 дюймов).

# Отображаем вектор v1.
# 0, 0 - координаты начала вектора (в данном случае - начало координат).
# *v1 - координаты конца вектора (распаковываются из кортежа v1).
# angles="xy" - указывает, что углы заданы в декартовой системе координат.
# scale_units="xy" - указывает, что единицы масштаба по осям x и y одинаковы.
# scale=1 - задает масштаб (в данном случае - без масштабирования).
# color="blue" - задает цвет вектора (синий).
# label="Вектор v1 (4, 3)" - задает подпись для легенды.
plt.quiver(0, 0, *v1, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор v1 (4, 3)")

# Аналогично отображаем вектор v2.
plt.quiver(0, 0, *v2, angles="xy", scale_units="xy", scale=1, color="green", label="Вектор v2 (-1, 1)")

# Аналогично отображаем сумму векторов v1 и v2.
plt.quiver(0, 0, *v_sum, angles="xy", scale_units="xy", scale=1, color="red", label="Сумма v1 + v2 (3, 4)")

# Настройка осей
# Задаем пределы отображения по осям x и y.
plt.xlim(-2, 5)
plt.ylim(-2, 5)

# Рисуем оси x и y.
plt.axhline(0, color="black",linewidth=0.5)
plt.axvline(0, color="black",linewidth=0.5)

# Включаем отображение сетки.
plt.grid(True)

# Делаем оси равномасштабными (чтобы векторы отображались корректно).
plt.gca().set_aspect('equal', adjustable='box')

# Отображаем легенду.
plt.legend()

# Добавляем заголовок к графику.
plt.title("Сложение векторов")

# Отображаем график.
plt.show()
```

```python
# Функция скалярного умножения
# Определяет функцию scale, которая принимает скаляр (число) и вектор в качестве аргументов.
def scale(scalar, v):
    # Возвращает новый вектор, полученный умножением исходного вектора на скаляр.
    # Умножение выполняется поэлементно: каждая координата вектора умножается на скаляр.
    return (scalar * v[0], scalar * v[1])

# Исходный вектор
# Определяет исходный вектор v.
v = (2, 3) # Вектор v имеет координаты (2, 3).
scaled_v = scale(2.5, v) # Вызывает функцию scale для умножения вектора v на скаляр 2.5 и сохраняет результат в переменной scaled_v.

# Построение векторов
# Создает новую фигуру для графика.
plt.figure(figsize=(6, 6)) # Задает размер фигуры (ширина 6 дюймов, высота 6 дюймов).

# Отображаем исходный вектор v.
# 0, 0 - координаты начала вектора (в данном случае - начало координат).
# *v - координаты конца вектора (распаковываются из кортежа v).
# angles="xy" - указывает, что углы заданы в декартовой системе координат.
# scale_units="xy" - указывает, что единицы масштаба по осям x и y одинаковы.
# scale=1 - задает масштаб (в данном случае - без масштабирования).
# color="blue" - задает цвет вектора (синий).
# label="Вектор v (2, 3)" - задает подпись для легенды.
plt.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор v (2, 3)")

# Отображаем вектор, полученный умножением на скаляр.
plt.quiver(0, 0, *scaled_v, angles="xy", scale_units="xy", scale=1, color="red", label="Скалярное произведение 2.5v (5, 7.5)")

# Настройка осей
# Задаем пределы отображения по осям x и y.
plt.xlim(-1, 6)
plt.ylim(-1, 8)

# Рисуем оси x и y.
plt.axhline(0, color="black",linewidth=0.5)
plt.axvline(0, color="black",linewidth=0.5)

# Включаем отображение сетки.
plt.grid(True)

# Делаем оси равномасштабными.
plt.gca().set_aspect('equal', adjustable='box')

# Отображаем легенду.
plt.legend()

# Добавляем заголовок к графику.
plt.title("Умножение вектора на скаляр")

# Отображаем график.
plt.show()
```

```python
from math import sqrt

# Функция для вычисления длины вектора
# Определяет функцию length, которая принимает вектор v в качестве аргумента.
def length(v):
    # Возвращает длину (или величину) вектора v.
    # Вычисляется как квадратный корень из суммы квадратов координат вектора.
    return sqrt(v[0]**2 + v[1]**2)

# Исходный вектор
# Определяет исходный вектор v.
v = (4, 3) # Вектор v имеет координаты (4, 3).
v_len = length(v) # Вызывает функцию length для вычисления длины вектора v и сохраняет результат в переменной v_len.

# Выводит длину вектора на экран.
# f-строка используется для форматирования вывода.
# {v} - подставляет значение переменной v (то есть координаты вектора).
# {v_len:.2f} - подставляет значение переменной v_len (то есть длину вектора),
#               форматируя его как число с плавающей точкой с двумя знаками после запятой.
print(f"Длина вектора {v}: {v_len:.2f}")
```

```python
# Функция для смещения точек на заданный вектор
def translate(translation, points):
    # Используем списковое включение для смещения каждой точки
    # Каждая точка 'p' состоит из координат (x, y)
    # Мы добавляем соответствующие значения вектора смещения 'translation' к координатам 'p'
    return [(p[0] + translation[0], p[1] + translation[1]) for p in points]

# Исходные точки, которые будут смещены
points = [(1, 2), (3, 4), (5, 6)]

# Вектор смещения, на который необходимо сдвинуть точки
translation = (-2, -1)

# Применяем функцию translate для получения смещённых точек
shifted_points = translate(translation, points)

# Вывод результата в читаемом формате
print(f"Исходные точки: {points}")  # Печатаем исходные координаты точек
print(f"Смещённые точки: {shifted_points}")  # Печатаем смещённые координаты точек
```

```python
# Функция для построения вершин параллелограмма по двум векторам
def parallelogram(v1, v2):
    # Возвращаем список координат вершин параллелограмма:
    # v1 - первая вершина (A),
    # add(v1, v2) - вторая вершина (B),
    # v2 - третья вершина (C),
    # (0, 0) - нулевая точка (D) для замыкания фигуры,
    # v1 - возвращаем первую вершину для завершения графического отображения.
    return [v1, add(v1, v2), v2, (0, 0), v1]

# Исходные векторы u и v, задающие параллелограмм
u = (3, 2)  # Вектор u
v = (1, 4)  # Вектор v

# Получение координат вершин параллелограмма
parallelogram_points = parallelogram(u, v)

# Построение графика
plt.figure(figsize=(6, 6))  # Установка размера графика
plt.plot(*zip(*parallelogram_points), color="orange", linestyle="-", marker="o", label="Параллелограмм")
# Построение параллелограмма по его вершинам с оранжевым цветом и маркерами

# Отображение векторов u и v с помощью функции quiver
plt.quiver(0, 0, *u, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор u")
plt.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1, color="green", label="Вектор v")

# Настройка осей графика
plt.xlim(-1, 6)  # Устанавливаем пределы по оси x
plt.ylim(-1, 6)  # Устанавливаем пределы по оси y
plt.axhline(0, color="black",linewidth=0.5)  # Добавляем ось x
plt.axvline(0, color="black",linewidth=0.5)  # Добавляем ось y
plt.grid(True)  # Включаем сетку на графике
plt.gca().set_aspect('equal', adjustable='box')  # Устанавливаем равные масштабы по осям

# Добавление легенды и заголовка
plt.legend()
plt.title("Параллелограмм по двум векторам")

# Отображение графика
plt.show()
```

```python
# Функция для вычитания двух векторов
def subtract(v1, v2):
    # Возвращает разность векторов v1 и v2, вычитая соответствующие координаты
    return (v1[0] - v2[0], v1[1] - v2[1])

# Исходные векторы, которые будут вычитаться
v1 = (5, 4)  # Первый вектор
v2 = (2, 3)  # Второй вектор

# Вычисление разности v1 - v2 с использованием функции subtract
v_diff = subtract(v1, v2)

# Начало визуализации результатов
plt.figure(figsize=(6, 6))  # Установка размера графика

# Отображение первого вектора v1
plt.quiver(0, 0, *v1, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор v1 (5, 4)")

# Отображение второго вектора v2
plt.quiver(0, 0, *v2, angles="xy", scale_units="xy", scale=1, color="green", label="Вектор v2 (2, 3)")

# Отображение разности v1 - v2
plt.quiver(0, 0, *v_diff, angles="xy", scale_units="xy", scale=1, color="red", label="Разность v1 - v2 (3, 1)")

# Настройка осей графика
plt.xlim(-1, 6)  # Установка предела по оси x
plt.ylim(-1, 6)  # Установка предела по оси y
plt.axhline(0, color="black", linewidth=0.5)  # Добавление горизонтальной оси
plt.axvline(0, color="black", linewidth=0.5)  # Добавление вертикальной оси
plt.grid(True)  # Включение сетки на графике

# Установка равных масштабов для осей
plt.gca().set_aspect('equal', adjustable='box')

# Добавление легенды и заголовка к графику
plt.legend()
plt.title("Вычитание векторов")

# Отображение графика
plt.show()
```

```python
# Импортируем функцию sqrt из модуля math для вычисления квадратного корня
from math import sqrt

# Функция для вычисления расстояния между двумя точками в 2D пространстве
def distance(p1, p2):
    # Вычисляем расстояние по формуле: sqrt((x2 - x1)^2 + (y2 - y1)^2)
    return sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

# Исходные точки, между которыми будет вычисляться расстояние
p1 = (1, 2)  # Первая точка с координатами (1, 2)
p2 = (4, 6)  # Вторая точка с координатами (4, 6)

# Вычисляем расстояние между точками p1 и p2
dist = distance(p1, p2)

# Печатаем результат с форматированием до двух знаков после запятой
print(f"Расстояние между точками {p1} и {p2}: {dist:.2f}")
```
